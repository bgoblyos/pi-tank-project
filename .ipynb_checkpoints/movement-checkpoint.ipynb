```python
import skimage
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from scipy.optimize import curve_fit
```

```python
right = pd.DataFrame(
    columns = ["time", "x1", "x2", "y"],
    data = [
        [0.05, 15, 25, 2],
        [0.06, 15, 25, 3.5],
        [0.07, 15, 25, 4],
        [0.08, 15, 25, 5.5],
        [0.09, 15, 25, 5.5],
        [0.10, 15, 25, 7],
        [0.11, 15, 25, 8.5],
        [0.12, 15, 25, 7.5],
        [0.13, 15, 25, 11],
    ]
)

right['angle'] = np.atan(right.y/right.x1)
right['angle_c'] = np.atan(right.y/right.x2)
right
```

```python
plt.scatter(right.angle_c, right.time)
plt.xlim(0, 0.5)
plt.ylim(0, 0.14)
```

```python
def model(x, a, b):
    sign = np.sign(x)
    x = x * sign
    return sign * (a*x + b*np.sqrt(x))

res, _ = curve_fit(model, right.angle_c, right.time)
plt.scatter(right.angle_c, right.time)
xs = np.linspace(0, 0.5, 100)
plt.plot(xs, model(xs, *res), color="orange")
plt.show()
res
```

```python
def getTime(tgt_angle):
    x = np.abs(tgt_angle)

    # Magic numbers (obtained with regression)
    a = 0.11596371
    b = 0.13241551

    # Use a linear + squareroot term to model the non-zero intercept
    # while still going smoothly to 0
    t = a*x + b*np.sqrt(x)
    
    # Return time and direction (True = right, False = left)
    return t, tgt_angle <= 0
```
