```python
from graphviz import Digraph
```

```python
# Create a directed graph
dot = Digraph(comment='Flowchart Example')# Add nodes
dot.node('start', 'Start')
dot.node('img', 'Take image')
dot.node('sign', 'Sign found?')
dot.node('perturb', 'Perturb heading')
dot.node('heading_check', 'Is sign centered?')
dot.node('turn', 'Turn to sign')
dot.node('close', 'Is sign close?')
dot.node('fwd', 'Move forward')
dot.node('img_rec', 'Recognise sign?')
dot.node('execute', 'Execute sign')

dot.edge('start', 'img')
dot.edge('img', 'sign')
dot.edge('sign', 'heading_check', label='Yes')
dot.edge('sign', 'perturb', label='No')
dot.edge('perturb', 'img')
dot.edge('heading_check', 'turn', label='No')
dot.edge('turn', 'img')
dot.edge('heading_check', 'close', label ='Yes')
dot.edge('close', 'fwd', label='No')
dot.edge('fwd', 'img')
dot.edge('close', 'img_rec', label ='Yes')
dot.edge('img_rec', 'execute', label='Yes')
dot.edge('execute', 'img')
dot.edge('img_rec', 'perturb', label='No')

dot.render()
dot
```
